```python
from sklearn.datasets import load_wine
wine = load_wine()
print(wine.DESCR)
```

```python
type(wine)
```

```python
import pandas as pd
```

```python
x = pd.DataFrame(wine.data, columns=wine.feature_names)
```

```python
x
```

```python
sy = pd.Series(wine.target, dtype="category")
```

```python
sy
```

```python
x.isnull().sum()
```

```python
df=x[:]
df['class'] = sy
df.tail()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
sns.pairplot(vars=df.columns[:-1], 
             hue="class", data=df)
plt.show()
```

```python
data=['flavanoids','color_intensity','proline','hue']
temp_x = x[data]
```

```python
sns.pairplot(vars=data, 
             hue="class", data=df)
plt.show()
```

```python
sns.kdeplot( hue=, data=df)
plt.show()
```

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(temp_x,sy,test_size = 0.2, random_state= 0)
```

```python
print(x_train.shape)
print(y_train.shape)
print('----')
print(x_test.shape)
print(y_test.shape)
```

```python
from sklearn import tree

model = tree.DecisionTreeClassifier(max_depth = 3, random_state = 0)

model.fit(x_train,y_train)

model.score(x_test,y_test)
```

```python

from sklearn.tree import plot_tree
plt.figure(dpi=300)
plot_tree(model,feature_names = x_train.columns,filled = True)
```
